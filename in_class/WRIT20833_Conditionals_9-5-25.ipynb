# In-Class Assignment: Conditional Flowchart Coding

> Conditional logic (if/else) isn’t a strange new way of thinking—it’s just codified common sense.  
>
> Every day you already make branching choices (*If it’s raining, bring an umbrella; else, don’t*).  
>
> Coding simply makes that reasoning formal and precise so a computer can carry it out quickly—and at a scale far beyond what we could manage ourselves.
